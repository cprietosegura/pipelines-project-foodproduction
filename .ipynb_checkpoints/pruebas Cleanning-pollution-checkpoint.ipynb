```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
df_emis = pd.read_csv("./output/Scrapped-Pollution-emissions2.csv")
```

```python
df_emis.head()
```

```python
def SelectCols(df):
    df=df[['Country', '1990', '1991', '1992', '1993', '1994', '1995', '1996',
       '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013']]
    return df
```

```python
df_emis=SelectCols(df_emis)
```

```python
df_emis.head()
```

```python
def CorrectNameCols(df):
    df['Country'][28]='Slovakia'
    return df
```

```python
df_emis=CorrectNameCols(df_emis)
```

```python
def SelectCountries(df):
    eu=['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Republic of Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Finland',
    'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 
    'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'United Kingdom']

    df = df[df['Country'].isin(eu)]
    return df
```

```python
df_emis=SelectCountries(df_emis)
```

```python
df_emis.head()
```

```python
def ReplaceUnknownVal(df):
    df=df.replace('..', 0.0)
    return df
 
```

```python
df_emis=ReplaceUnknownVal(df_emis)
```

```python
df_emis
```

```python
def ChangeColType(df):
    for col in df.columns:
        if col=='Country':
            df[col] = df[col].astype(str)
        else:
            df[col] = df[col].astype(float)
    return df
```

```python
df_emis=ChangeColType(df_emis)
```

```python
def TransformDf(df):
    df=df.T
    df.columns = df.iloc[0]
    df.drop(df.tail(1).index,inplace=True) 
    df.drop(df.head(1).index,inplace=True)
    df.reset_index(level=0, inplace=True)
    df = df.rename(columns={"index": "Year"})
    
    return df
```

```python
df_emis=TransformDf(df_emis)
```
