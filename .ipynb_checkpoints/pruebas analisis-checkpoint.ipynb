```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def OpenData(file):
    df = pd.read_csv(file)
    return df
```

```python
dffood=OpenData(r'./output/food-production.csv')
dfpoll=OpenData(r'./output/Pollution-emissions.csv')
```

```python
dfpoll.head()
```

```python
dffood.head()
```

```python
def Maximum(df_food,df_poll,country): #hay que crear la excepción por si el país no está.
        max_year_food = df_food.loc[df_food[country].idxmax()]
        max_year_poll = df_poll.loc[df_poll[country].idxmax()]
        return 'The year with more food production was {} and with more gas emissions was {}.'.format(int(max_year_food[1]), int(max_year_poll[1]))
```

```python
maximos=Maximum(dffood,dfpoll,'United States')
    
```

```python
maximos
```

```python
def CompareEvolutionGraffics(df_food,df_poll,country):
    grafica_1=df_food.plot(x='Year', y=[country])
    grafica_2=df_poll.plot(x='Year', y=[country])
    return grafica_1,grafica_2
```

```python
grafica_1, grafica_2 = CompareEvolutionGraffics(dffood,dfpoll,'France')
```

```python
def SaveGraffics(grafica_1, grafica_2): # TENGO DOS GRAFICAS
    fig = grafica_1.get_figure()
    fig.savefig('graf_food' + '.png')
    fig = grafica_2.get_figure()
    fig.savefig('graf_poll' + '.png')
```

```python
SaveGraffics(grafica_1, grafica_2)
```
