```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
def OpenData(file):
    df = pd.read_csv(file)
    return df
```

```python
dffood=OpenData(r'./output/food-production.csv')
dfpoll=OpenData(r'./output/Pollution-emissions.csv')
```

```python
def Maximum(df,country):
    max_year = df.loc[df[country].idxmax()]
    return 'The year with more production was:',int(max_year[1])
```

```python
maximos=Maximum(dffood,'France')
    
```

```python
def CompareEvolutionGraffics(df_food,df_poll,country):
    grafica_1=df_food.plot(x='Year', y=[country])
    grafica_2=df_poll.plot(x='Year', y=[country])
    return grafica_1,grafica_2
```

```python
CompareEvolutionGraffics(dffood,dfpoll,'Spain')
```

```python
grafica_1, grafica_2 = CompareEvolutionGraffics(dffood,dfpoll,'Spain')
```

```python
def SaveGraffics(grafica_1, grafica_2): # TENGO DOS GRAFICAS
    fig = grafica_1.get_figure()
    fig.savefig('graf_food' + '.png')
    fig = grafica_2.get_figure()
    fig.savefig('graf_poll' + '.png')
```

```python
SaveGraffics(grafica_1, grafica_2)
```
